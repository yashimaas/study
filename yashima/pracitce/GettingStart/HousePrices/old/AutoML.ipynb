```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
pd.set_option('display.max_columns', 90)
```

```python
'''
データのロード
'''
import pickle

pickle_file = "./tuned_data3.pickle"
with open(pickle_file, 'rb') as f:
    X_train = pickle.load(f)
    Y_train = pickle.load(f)
    X_test = pickle.load(f)
```

```python
'''
特徴量の選択
'''
from boruta_py import BorutaPy
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(max_depth=3)
feat_selector = BorutaPy(
    model,
    n_estimators='auto',
    two_step= True,
    verbose=2,
    random_state=42,
    perc=80,
    max_iter=150
)
```

```python
_ = pd.concat([X_train, Y_train], axis=1)
_.dropna(inplace=True)
X_train_droped = _.drop(["SalePrice"], axis=1)
Y_train_droped = _.SalePrice
X_train_np = np.array(X_train_droped)
Y_train_np = np.array(Y_train_droped)
X_test_np = np.array(X_test)
```

```python
feat_selector.fit(X_train_np, Y_train_np)
```

```python
print(X_train_np.shape)
print(Y_train_np.shape)
print(X_test_np.shape)
```

```python
X_train_np = X_train_np[:,feat_selector.support_]
X_test_np = X_test_np[:,feat_selector.support_]
```

```python
'''
データの保存
'''
import pickle

Id = test_data.Id

pickle_file = "./droped29_HousePrice_data_kernel.pickle"
with open(pickle_file, 'wb') as f:
    pickle.dump(X_train, f)
    pickle.dump(Y_train, f)
    pickle.dump(X_test, f)
    pickle.dump(train_data, f)
    pickle.dump(test_data, f)
    pickle.dump(Id, f)
```

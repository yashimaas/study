# Introduction

Lets first discuss what we are given and what we have to predict. About our dataset :

We have in our training data :
1. date - every date of items sold
2. date_block_num - this number given to every month
3. shop_id - unique number of every shop
4. item_id - unique number of every item
5. item_price - price of every item
6. item_cnt_day - number of items sold on a particular day 

We have inour testing data :
1. ID - unique for every (shop_id,item_id) pair.
2. shop_id - unique number of every shop
3. item_id - unique number of every item

**Now what we have to predict ?** \
We have to predict <font color="Red">how many items of a type from each shop will be sold in a whole month. </font> \
Our submission should have ID and item_cnt_month columns.

**What is our approach?** \
Our approach will be simple. \
Our features will be number of items sold in month from a shop excluding last month data because that will our labels, that we help our model learn to predict next sequence. \
And for testing will use number of items sold in month from a shop excluding first month like this dimension of our data remains same. \
<font color="Red">Our model will predict the next sequence and that we will be our results.</font> \
This is pretty simple approach but its good for start. Please try some different approaches also.

# Preprocess

## Convert 'date' on DataFrame format

```python
sales_data['date'] = pd.to_datetime(sales_data['date'],format = '%d.%m.%Y')
```

```python
sales_data.head(3)
```

## Make Pivot Table

Now we will create a [pivot tabel][1] by going so we get our data in desired form. \
We want get total <font color="Red">count value of an item over the whole month for a shop</font>. \
That why we made <u>shop_id and item_id our indices</u> and <u>date_block_num our column</u>. \
The value we want is item_cnt_day and used sum as aggregating function. 

[1]:https://note.nkmk.me/python-pandas-pivot-table/

Note : 
- shop_id - unique identifier of a shop
- item_id - unique identifier of a product
- date_block_num - a consecutive month number, used for convenience. January 2013 is 0, February 2013 is 1,..., October 2015 is 33

```python
dataset = sales_data.pivot_table(
    index = ['shop_id','item_id'],
    values = ['item_cnt_day'],
    columns = ['date_block_num'],
    fill_value = 0,
    aggfunc='sum') # <- default : 'mean'
dataset.reset_index(inplace = True)
```

```python
dataset.head(3)
```

Now we will <u>merge our pivot table with the test_data</u> because we want to keep the data of items we have for predict.

```python
test_data.head(3)
```

```python
dataset = pd.merge(test_data,dataset,on = ['item_id','shop_id'],how = 'left')
dataset.fillna(0,inplace = True)
```

```python
dataset.head(3)
```

We will <u>drop shop_id and item_id</u> because we do not need them for predict.\
i.e. We are teaching our model <font color='Red'>how to generate the next sequence</font> and it will be enough to do only Index.

```python
dataset.drop(['shop_id','item_id','ID'],inplace = True, axis = 1)
```

```python
dataset.head(3)
```

## Split Data

```python
x = [1,2,3,4,5]
x[-1:]
```

```python
# X we will keep all columns execpt the last one 
X_train = np.expand_dims(dataset.values[:,:-1],axis = 2)
# the last column is our label
y_train = dataset.values[:,-1:]
# for test we keep all the columns execpt the first one
X_test = np.expand_dims(dataset.values[:,1:],axis = 2)
```

```python
print(X_train.shape,y_train.shape,X_test.shape)
```

# Modeling

## Make model

```python
model = Sequential()
model.add(LSTM(units = 64,input_shape = (33,1)))
model.add(Dropout(0.4))
model.add(Dense(1))

model.compile(
    loss = 'mse',
    optimizer = 'adam', 
    metrics = ['mean_squared_error'])
```

```python
model.summary()
```

## Fit

```python
callbacks = [
    keras.callbacks.TensorBoard(log_dir='../logs/log_kernel_LSTM', histogram_freq=1)
]

### Usage ###
# 
# Do below command in your command line if you want to check learning process.
#
# $ tensorboard --logdir=/full_path_to_your_logs
# 
# or if you use tensorbord on jupter notebook, check "7.References Tensorbord on JupyterNotebook".
```

```python
model.fit(
    X_train,y_train,
    batch_size = 4096,
    epochs = 10,
    callbacks=callbacks)
```

# Submit

```python
pred = model.predict(X_test)
pred = pred.clip(0,20)
submission = pd.DataFrame({'ID':test_data['ID'],'item_cnt_month':pred.ravel()})
submission.to_csv('../submissions/submission_template.csv',index = False)
```
